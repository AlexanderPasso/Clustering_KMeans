### Librerias utilizadas 

```python
#Librerias 

import numpy as np
import matplotlib.pyplot as plt 
import random                          # crear numeros aleatoriamente
import scipy.io as sio
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
import pandas as pd
from yellowbrick.cluster import KElbowVisualizer   #Permite encontrar el mejor numero de cluster (K) a utilizar.
                                                   #Y retorna la grafica de k vs distorsion
```

### Definición de funciones para el pre-procesamiento de los datos

```python
# Función que permite la extraccion de datos de entrenamiento, prueba y validacion
# para una base de datos .mat
def datos(data):
    """
    data: es el archivo.mat para leer todos los datos
    D_entre: Salida que retorna la cantidad de datos de entrenamiento 
    D_prueba: Salida que retorna la cantidad de datos de prueba
    D_val: Salida que retorna la cantidad de datos de Validacion
    """
    
    #Extrayendo matriz de caracteristicas 
    Xi = data['x']

    #Extrayendo dato esperado (Clases)
    Yi = data['y']

    #Agregando el bias a las filas de la matriz de caracteristicas (Entradas)
    #bias = np.full((Xi.shape[0], 1), 1)
    #Xi = np.hstack((bias, Xi))

    #numero de datos
    num_datos = Xi.shape[0]

    #numero de caracteristicas
    caract = Xi.shape[1]              

    #Datos de entrenamiento y prueba: entrenamiento 90% , prueba 10%
    X_ent, X_pru, Y_ent, Y_pru = train_test_split(Xi, Yi, test_size=0.1)

    #Obteniendo datos de validacion: Validacion 10% y Entrenamiento 80%
    #Se extraen los datos de validación a partir de los de entrenamiento
    X_ent, X_val, Y_ent, Y_val = train_test_split(X_ent, Y_ent, test_size=0.1, random_state=42)

    #Obteniendo datos de entrenamiento y prueba
    D_entre  =  len(X_ent)       # datos de entrenamiento   (verificar la cantidad de dtos para hacer la particion)
    D_prueba =  len(X_pru)       #Datos de validacion 
    D_val    =  len(X_val)

    print("Cantidad de datos del dataset: ", Xi.shape)
    print("Cantidad de datos de entrenamiento:",D_entre)
    print("Cantidad de datos de prueba:",D_prueba)
    print("Cantidad de datos de Validación:", D_val)
    
    #se grafica la distribución de los datos
    plot_datos(Xi,Yi,X_ent,Y_ent,X_pru,Y_pru,X_val,Y_val)
    
    return Xi,Yi,X_ent, Y_ent, D_entre,X_pru,Y_pru, D_prueba,X_val, Y_val, D_val
```

```python
def plot_datos(Xi,Yi,X_ent,Y_ent,X_pru,Y_pru,X_val,Y_val):
    
    # Crear la figura y los subplots
    fig, axs = plt.subplots(2, 2, figsize=(14, 14))

    # Graficar la segunda gráfica en el segundo subplot
    axs[0, 0].scatter(Xi[:, 0], Xi[:, 1], c=Yi, cmap=plt.cm.coolwarm)
    axs[0, 0].set_title('Datos Originales', fontsize=18)
    axs[0, 0].set_xlabel('X1', fontsize=14)
    axs[0, 0].set_ylabel('X2', fontsize=14)

    # Graficar la primera gráfica en el primer subplot
    axs[0, 1].scatter(X_ent[:, 0], X_ent[:, 1], c=Y_ent, cmap=plt.cm.coolwarm)
    axs[0, 1].set_title('Datos de Entrenamiento', fontsize=18)
    axs[0, 1].set_xlabel('X1', fontsize=14)
    axs[0, 1].set_ylabel('X2', fontsize=14)

    # Graficar la tercera gráfica en el tercer subplot
    axs[1, 0].scatter(X_pru[:, 0], X_pru[:, 1], c=Y_pru, cmap=plt.cm.coolwarm)
    axs[1, 0].set_title('Datos de Prueba', fontsize=18)
    axs[1, 0].set_xlabel('X1', fontsize=14)
    axs[1, 0].set_ylabel('X2', fontsize=14)

    # Graficar la cuarta gráfica en el cuarto subplot
    axs[1, 1].scatter(X_val[:, 0], X_val[:, 1], c=Y_val, cmap=plt.cm.coolwarm)
    axs[1, 1].set_title('Datos de Validación', fontsize=18)
    axs[1, 1].set_xlabel('X1', fontsize=14)
    axs[1, 1].set_ylabel('X2', fontsize=14)

    # Mostrar la figura
    plt.show()
```

```python
#Funcion para normalizar los datos

def standard_function(df_input):
    return (df_input - df_input.min()) / ( df_input.max() - df_input.min())
```

## Dataset: TiosulfatoSodio

```python
data = sio.loadmat('Tiosulfatosodio.mat') # cargar el archivo .mat
Xi,Yi,X_ent, Y_ent, D_entre,X_pru,Y_pru, D_prueba,X_val, Y_val, D_val = datos(data)
```

### Conversión de datos a un DataFrame

```python
#Crear un pandas con los datos Xi - Yi 
df = pd.DataFrame()

df['X1'] = Xi[:,0]
df['X2'] = Xi[:,1]
df['Y']  = Yi[:,]

df_normal = standard_function(df)

#Datos de entrenamiento y prueba: entrenamiento 90% , prueba 10%
X_ent, X_pru, Y_ent, Y_pru = train_test_split(df_normal.iloc[:, :2], df_normal.iloc[:,-1], test_size=0.1)

#Obteniendo datos de validacion: Validacion 10% y Entrenamiento 80%
#Se extraen los datos de validación a partir de los de entrenamiento
X_ent, X_val, Y_ent, Y_val = train_test_split(X_ent, Y_ent, test_size=0.1, random_state=42)
```

### Kmeans utilizando la libreria Sklearn
#### Se utiliza la libreria yellowbrick que permite obtener la grafica del 'codo', que relaciona la cantidad de cluster y el puntaje de distorsión. 

```python
#Se aplica kmeans utilizando la libreria Sklearn
#Parametros de la funcion kmeans
"""
X,
n_clusters,
*,
sample_weight=None,
init="k-means++",
n_init="warn",
max_iter=300,
verbose=False,
tol=1e-4,
random_state=None,
copy_x=True,
algorithm="lloyd",
return_n_iter=False
"""

    
#Mostramos los clusters recomendados 
kmeans = KMeans()

visualizer = KElbowVisualizer(kmeans,  k=(1,12))
visualizer.fit(X_ent)
visualizer.show() 
```

### Entrenando el modelo con el algoritmo Kmeans
#### Se Crea un nuevo DataFrame para la organización de los datos de entrenamiento

```python
kmeans = KMeans(n_clusters = 10, init="k-means++", max_iter= D_entre)
kmeans.fit(X_ent)


#print(kmeans.labels_[:20])
dfCluster = pd.DataFrame()
# agregar etiquetas de cluster al dataframe
dfCluster['X1_ent'] = X_ent['X1']
dfCluster['X2_ent'] = X_ent['X2']
dfCluster['Y_ent'] = Y_ent[:,]
dfCluster['cluster'] = kmeans.labels_
dfCluster
```

### Cantidad de '1s' y '0s' en cada cluster.

```python

# contar número de "0" y "1" en cada cluster
cluster_counts = dfCluster.groupby(['cluster', 'Y_ent']).size().unstack(fill_value=0)
print(cluster_counts)
```

### Gráfica de los Cluster y obtención de sus centros.

```python
#graficar clusters
plt.scatter(dfCluster['X1_ent'], dfCluster['X2_ent'], c=kmeans.labels_ , cmap=plt.cm.Dark2)
centros = kmeans.cluster_centers_

#Mostrar los centros de cada clusters
for i in range(len(centros)):
    #print(centros[i])
    x1 = np.round(centros[i][0],4)
    y1 = np.round(centros[i][1],4)
    print("Centro cluster ", i , ": ", "(" , x1, "," , y1, ")")
    plt.plot(x1,y1, 'X' ,color = 'black', markersize=15)
```

### Condición de correspondencia para los datos de entrenamiento 
##### Se tiene en cuenta la cantidad de '1s' y '0s' que hay en cada cluster, utilizando los datos reales y sus etiquetas. 

```python
#Condicion de correspondencia

#creo una nueva columna para agregar la etiqueta a cada cluster segun la cantidad de 1 o -1 
dfCluster['Ys'] = 0

# Recorro los cluster
for i in range(cluster_counts.shape[0]):
    if(cluster_counts.iloc[i][0]> cluster_counts.iloc[i][1]):   #Pregunto si hay mas ceros que unos en cada cluster
        print("Cluster: ", i , "Hay mas ceros")
        
        for j in range(dfCluster['cluster'].shape[0]):   #Recorro todas las posiciones de la columna cluster 
            if (dfCluster['cluster'].values[j] == i ):#Encuentro las posiciones donde hay coincidencia y agg cero a la nueva columna Ys
                dfCluster['Ys'].values[j] = 0
        
    else:
        print("Cluster: ", i ,"hay mas unos")          #Si se cumple esta condicion es porque hay mas unos en el cluster actual
        for j in range(dfCluster['cluster'].shape[0]): #Recorro todas las posiciones del cluster
            if (dfCluster['cluster'].values[j] == i ): #Encuentro las posiciones donde hay coincidencia y agg 0 a la columna Ys
                dfCluster['Ys'].values[j] = 1

#Muestra el nuevo dataframe donde se guarda los datos de entrenamiento, clusters, etiquetas a partir de la correspondencia(Ys)
# Y salida(Y) dada por la cantidad de '1' o '0' en cada clusters
dfCluster
```

### Calculo del error de entrenamiento

```python
error_ent = np.where(dfCluster['Y_ent'] == dfCluster['Ys'], 0, 1)
E_porc_ent = np.round(((error_ent.sum())/(D_entre))*100 ,3)
print("Error de entrenamiento: ", E_porc_ent , "%")
```

### Análisis para los datos de prueba

### Calculo de las distancias de cada dato de prueba a los centros de los cluster a través de la distancia euclidiana para determinar a que cluster pertenece cada dato a partir de la distancia minima

```python
#Datos de prueba 
#Calculo de la distancia de cada punto a los centros de los cluster
#Y asignar '1' o '0'

distancias_minimas = []
clusters_prueba = []

for i in range(D_prueba):
    distancia_minima = float('inf')
    cluster_actual = None
    for j in range(cluster_counts.shape[0]):
        distancia = np.sqrt( ( centros[j][0] - X_pru['X1'].values[i])**2 + ( centros[j][1] - X_pru['X2'].values[i] )**2 )
        if distancia < distancia_minima:
            distancia_minima = distancia
            cluster_actual = j
    distancias_minimas.append(distancia_minima)
    clusters_prueba.append(cluster_actual)
    #imprimo dato, cluster mas cercado, y distancia al cluster mas cercano
    #print("Dato: ", i, "cluster más cercano: ", cluster_actual, "distancia mínima: ", distancia_minima)
    

#Creo un nuevo dataframe para organizar los datos de prueba 
dfPrueba = pd.DataFrame()
dfPrueba['X1_pru'] = X_pru['X1']
dfPrueba['X2_pru'] = X_pru['X2']
dfPrueba['Y_pru' ] = Y_pru[:]
dfPrueba['cluster_pru'] = clusters_prueba
```

### Contar número de "0" y "1" en cada cluster

```python
cluster_counts_pru = dfPrueba.groupby(['cluster_pru', 'Y_pru']).size().unstack(fill_value=0)
print(cluster_counts_pru)
```

```python

#creo una nueva columna para agregar la etiqueta a cada cluster segun la cantidad de 1 o -1 
dfPrueba['Ys'] = 0

# Recorro los cluster
for i in range(cluster_counts_pru.shape[0]):
    if(cluster_counts_pru.iloc[i][0]> cluster_counts_pru.iloc[i][1]):   #Pregunto si hay mas ceros que unos en cada cluster
        print("Cluster: ", i , "Hay mas ceros")
        
        for j in range(dfPrueba['cluster_pru'].shape[0]):   #Recorro todas las posiciones de la columna cluster 
            if (dfPrueba['cluster_pru'].values[j] == i ):#Encuentro las posiciones donde hay coincidencia y agg cero a la nueva columna Ys
                dfPrueba['Ys'].values[j] = 0
        
    else:
        print("Cluster: ", i ,"hay mas unos")          #Si se cumple esta condicion es porque hay mas unos en el cluster actual
        for j in range(dfPrueba['cluster_pru'].shape[0]): #Recorro todas las posiciones del cluster
            if (dfPrueba['cluster_pru'].values[j] == i ): #Encuentro las posiciones donde hay coincidencia y agg 0 a la columna Ys
                dfPrueba['Ys'].values[j] = 1

#Muestra el nuevo dataframe donde se guarda los datos de entrenamiento, clusters, etiquetas a partir de la correspondencia(Ys)
# Y salida(Y) dada por la cantidad de '1' o '0' en cada clusters

```

### Calculo del error de prueba 

```python
error_pru = np.where(dfPrueba['Y_pru'] == dfPrueba['Ys'], 0, 1)
E_porc_pru = np.round(((error_pru .sum())/(D_prueba))*100 ,3)
print("Error de prueba: ", E_porc_pru , "%")
```

### Análisis para los datos de validación 

```python
#Datos de validacion
#Calculo de la distancia de cada punto a los centros de los cluster
#Y asignar '1' o '0'

distancias_minimas = []
clusters_val = []

for i in range(D_val):
    distancia_minima = float('inf')
    cluster_actual = None
    for j in range(cluster_counts.shape[0]):
        distancia = np.sqrt( ( centros[j][0] - X_val['X1'].values[i])**2 + ( centros[j][1] - X_val['X2'].values[i] )**2 )
        if distancia < distancia_minima:
            distancia_minima = distancia
            cluster_actual = j
    distancias_minimas.append(distancia_minima)
    clusters_val.append(cluster_actual)
    #imprimo dato, cluster mas cercado, y distancia al cluster mas cercano
    #print("Dato: ", i, "cluster más cercano: ", cluster_actual, "distancia mínima: ", distancia_minima)
    

#Creo un nuevo dataframe para organizar los datos de prueba 
dfVal = pd.DataFrame()
dfVal['X1_val'] = X_val['X1']
dfVal['X2_val'] = X_val['X2']
dfVal['Y_val' ] = Y_val[:]
dfVal['cluster_val'] = clusters_val
```

```python
cluster_counts_val = dfVal.groupby(['cluster_val', 'Y_val']).size().unstack(fill_value=0)
print(cluster_counts_val)
```

```python

#creo una nueva columna para agregar la etiqueta a cada cluster segun la cantidad de 1 o -1 
dfVal['Ys'] = 0

# Recorro los cluster
for i in range(cluster_counts_val.shape[0]):
    if(cluster_counts_val.iloc[i][0]> cluster_counts_val.iloc[i][1]):   #Pregunto si hay mas ceros que unos en cada cluster
        print("Cluster ", i , " : ","Hay mas ceros")
        
        for j in range(dfVal['cluster_val'].shape[0]):   #Recorro todas las posiciones de la columna cluster 
            if (dfVal['cluster_val'].values[j] == i ):#Encuentro las posiciones donde hay coincidencia y agg cero a la nueva columna Ys
                dfVal['Ys'].values[j] = 0
        
    else:
        print("Cluster ", i , " : ","hay mas unos")          #Si se cumple esta condicion es porque hay mas unos en el cluster actual
        for j in range(dfVal['cluster_val'].shape[0]): #Recorro todas las posiciones del cluster
            if (dfVal['cluster_val'].values[j] == i ): #Encuentro las posiciones donde hay coincidencia y agg 0 a la columna Ys
                dfVal['Ys'].values[j] = 1

#Muestra el nuevo dataframe donde se guarda los datos de entrenamiento, clusters, etiquetas a partir de la correspondencia(Ys)
# Y salida(Y) dada por la cantidad de '1' o '0' en cada clusters
#dfVal     #Descomentar dfVal para visualizar el dataFrame de los datos de validacion con sus nuevas etiquetas
```
